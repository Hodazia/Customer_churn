```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('processed.csv')
```

```python
df.head(10)
```

```python
# drop the first unnamed :0 col
df.drop(columns='Unnamed: 0',inplace=True)
#df.columns
```

```python
df
```

```python
col = ['tenure','MonthlyCharges','TotalCharges']
# create a Column Transfer to transform these columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
transform = ColumnTransformer([
    ('scaler',StandardScaler(),col)
],
remainder='passthrough')
```

```python
X = df.drop(columns = ['Churn'])
y = df[['Churn']]

x_transform=transform.fit_transform(X)
```

```python
columns = transform.get_feature_names_out()
columns = list(map(lambda x: str(x).split("__")[-1], columns))
x_transform = pd.DataFrame(x_transform, columns = columns)
x_transform.head()
```

```python
# now you can split the dataset into train and test

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_transform,y,test_size=.2,random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,auc,classification_report,f1_score,precision_score,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV,KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# create a function for error metrics

def accuracy_metrics(y_train,y_test,y_train_pred,y_test_pred):

    acc={}
    acc['train_accuracy_score']=accuracy_score(y_train_pred,y_train)
    acc['train_f1_score'] = f1_score(y_train_pred,y_train)
    acc['train_precision'] = precision_score(y_train_pred,y_train)

    acc['test_accuracy_score']=accuracy_score(y_test_pred,y_test)
    acc['test_f1_score'] = f1_score(y_test_pred,y_test)
    acc['test_precision'] = precision_score(y_test_pred,y_test)

    return acc
```

```python
## Model Building
#1. KNN

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train,y_train)

y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)

accuracy_metrics(y_train,y_test,y_train_pred,y_test_pred)
```

```python
class_report = classification_report(y_test, y_test_pred)
print("\nClassification Report:")
print(class_report)

conf_matrix = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
# 2. KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_train_pred = knn.predict(x_train)
y_test_pred = knn.predict(x_test)

accuracy_metrics(y_train,y_test,y_train_pred,y_test_pred)
```

```python
# Hyperparameter tuning
kfold = KFold(n_splits=5,shuffle=True,random_state=42)
params = {
    'n_neighbors':[3,5,7,9,10,12],
    'weights':['uniform','distance'],
    'algorithm':['ball_tree','brute'],
    'p':[1,2]
}

gcv = GridSearchCV(knn,param_grid=params,cv=kfold,scoring='accuracy',verbose=2,refit=True)
```

```python
gcv.fit(x_train,np.array(y_train).reshape(y_train.shape[0],))
```

```python
gcv.best_params_,gcv.best_score_
```

```python
y_train_pred = gcv.predict(x_train)
y_test_pred = gcv.predict(x_test)

pd.Series(accuracy_metrics(y_train,y_test,y_train_pred,y_test_pred))
```

```python
class_report = classification_report(y_test, y_test_pred)
print("\nClassification Report:")
print(class_report)

conf_matrix = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
from imblearn.combine import SMOTEENN

# Apply SMOTEN to the training data
smoten = SMOTEENN(random_state=42)
X_resampled, y_resampled = smoten.fit_resample(X, y)
# Print the shapes after applying SMOTEN
print('X_resampled shape is ', X_resampled.shape)
print('y_resampled shape is ', y_resampled.shape)
```

```python
transform = ColumnTransformer([
    ('scaler',StandardScaler(),col)
],
remainder='passthrough')

x_transform=transform.fit_transform(X_resampled)


columns = transform.get_feature_names_out()
columns = list(map(lambda x: str(x).split("__")[-1], columns))
x_transform = pd.DataFrame(x_transform, columns = columns)
print(x_transform.head())


```

```python
X_train_resampled, X_test_resampled, y_train_resampled, y_test_resampled = train_test_split(x_transform, y_resampled,
                                                                           test_size=0.20,random_state=42, shuffle =True,stratify=y_resampled)
print('X_train_resampled shape is ' , X_train_resampled.shape)
print('X_test_resampled shape is ' , X_test_resampled.shape)
print('y_train_resampled shape is ' , y_train_resampled.shape)
print('y_test_resampled shape is ' , y_test_resampled.shape)
```

```python
X_train_resampled
```

```python
models = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'KNN': KNeighborsClassifier(),
    'SVC': SVC(random_state=42, class_weight='balanced'),
    'LogisticRegression': LogisticRegression(random_state=42, class_weight='balanced'),
}
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
# Apply ML in SMOTEN to the training data
### Validation Score
cv_results = {}
for model_name, model in models.items():
    print(f"Cross-validating {model_name}...")
    cv_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=5, scoring='accuracy', n_jobs=-1,verbose=0)
    cv_results[model_name] = cv_scores.mean()
    print(f"Mean accuracy for {model_name}: {cv_scores.mean():.4f}")
```

```python
cv_results_df = pd.DataFrame(list(cv_results.items()), columns=['Model', 'Mean Accuracy'])
cv_results_df = cv_results_df.sort_values(by='Mean Accuracy', ascending=False)
cv_results_df
```
